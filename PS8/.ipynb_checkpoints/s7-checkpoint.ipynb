```python
import numpy as np, matplotlib.pyplot as plt
%matplotlib inline

import torch
from torch import nn
from torchvision import transforms, datasets
from torch.utils.data import DataLoader
from collections import defaultdict
```

```python
# saved in s7_supp.ipynb
(X_train2k_scaled, X_test2k_scaled, X_train4k_scaled, X_test4k_scaled,
                        y_train, y_test) = np.load('auths20.npy', allow_pickle=True)
```

```python
X2k_train = torch.Tensor(X_train2k_scaled)
X2k_test = torch.Tensor(X_test2k_scaled)
X4k_train = torch.Tensor(X_train4k_scaled)
X4k_test = torch.Tensor(X_test4k_scaled)
y_train = torch.tensor(y_train)
y_test = torch.tensor(y_test)

X2k_train.shape,X2k_test.shape, X4k_train.shape,X4k_test.shape, y_train.shape,y_test.shape
```

```python
model = nn.Sequential(

  nn.Linear(2000,1024),
  nn.ReLU(), 
  nn.Dropout(.75),
    
  nn.Linear(1024,20),
  nn.Dropout(.75),
)

learning_rate = 1e-3;
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
for t in range(601):
    model.train()
    y_pred = model(X2k_train)
    loss = loss_fn(y_pred, y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    model.eval()
    tr_score = (model(X2k_train).argmax(1) == y_train).sum()/len(y_train)
    te_score = (model(X2k_test).argmax(1) == y_test).sum()/len(y_test)
    if te_score > .85: print('  ', t, te_score.item())
    if t%100==0: print(f'{t}, {loss.item():.5f}, {tr_score.item():.1%}, {te_score.item():.1%}')
```

```python
model = nn.Sequential(

  nn.Linear(4000,2048), 
  nn.ReLU(),
  nn.Dropout(.75),
    
  nn.Linear(2048,20),   # connects to 20 node output layer
  nn.Dropout(.75),
)

learning_rate = 1e-3;
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
for t in range(201):
    model.train()
    y_pred = model(X4k_train)
    loss = loss_fn(y_pred, y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    model.eval()
    tr_score = (model(X4k_train).argmax(1) == y_train).sum()/len(y_train)
    te_score = (model(X4k_test).argmax(1) == y_test).sum()/len(y_test)
    if te_score > .87: print('  ', t, te_score.item())
    if t%100==0: print(f'{t}, {loss.item():.5f}, {tr_score.item():.1%}, {te_score.item():.1%}')
```

```python
model99 = nn.Sequential(
    
    nn.Conv2d(1, 32, kernel_size=5),
    nn.ReLU(),
    nn.MaxPool2d(kernel_size=2),
    
    nn.Conv2d(32, 64, kernel_size=5),
    nn.ReLU(),
    nn.MaxPool2d(kernel_size=(2, 2)),
    
    nn.Flatten(),    
    nn.Linear(64*4*4, 10),
)
```

```python
model99
```

```python
#find total number of parameters
for name,layer in model99.named_parameters():
    print(f'{name:<10} {layer.shape:} {np.prod(layer.shape)}')
print('Total =', sum([np.prod(layer.shape) for layer in model99.parameters()]))
```

**A.** The ReLU, MaxPool, and Flatten layers have no parameters.

The first Conv2d layer learns 32 5x5 filters from a single channel of data, has `5*5=25` weights for each filter,  so has `32*1*5*5 = 800` weights. Each filter also has an overall additive bias term, giving the 32 more parameters in the second line above.

The second Conv2d layer learns 64 5x5 filters from 32 channels, so each filter has `5*5*32=800` weights, and 64 filters gives `64*800=51200` total weights (third line), and as well 64 biases (fourth line)

The final Linear layer has 10 outputs. Each output is a linear combination of 1024 input features plus an additive bias, so over all there are 10240 weights (fifth line above) and 10 biases (sixth line above).

**B.**

i) `nn.Linear(28*28,128)` layer has `28*28*128=100352` weights and 128 biases, `nn.Linear(128,10)` has `128*10 = 1280` weights and 10 biases, total is 100352 + 128 + 1280 + 10 = 62346 parameters

ii)  `nn.Linear(2000,1024)` layer has `2000*1024 = 2048000` weights and 1024 biases, `nn.Linear(1024,20)` layer has `1024*20 = 20480` weights and 20 biases, for a total of 2048000 + 1024 + 20480 + 20 = 2069524 parameters

iii) `nn.Linear(4000,2048)` layer has `4000*2048 = 8192000` weights and 2048 biases, `nn.Linear(2048,20)` layer has `2048*20 = 40960` weights and 20 biases, for a total of 8192000 + 2048 + 40960 + 20 = 8235028 parameters

For comparison, the 99.5% model IV has many more layers than ii,iii but only a total of 326410 parameters, because the initial convolutional layers train filters with relatively few parameters, so process the inputs more efficiently and are able to pass a smaller number of composite features to the later layers. (The largest single layer in that model is the `nn.Linear(64*4*4, 128)` with 131072 weights and 128 biases.) This suggests that there are as well more efficient models for text data, analogous to convolutional layers, but taking advantage of analogous properties of text data.

```python
train_data = datasets.MNIST(
    root="./",
    train=True,
#    download=True,
    transform=transforms.ToTensor()
)

# Test data
test_data = datasets.MNIST(
    root="./",
    train=False,
#    download=True, #already downloaded with train data above
    transform=transforms.ToTensor()
)
```

```python
# Dataloaders
batch_size = 64 #32 #128

X_train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
X_test_loader = DataLoader(test_data, batch_size=batch_size)
```

```python
def train_model(model, scores, epochs=15):
  optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
  for t in range(epochs):
    train_loss = []
    train_accuracy = []

    model.train() #turns on dropout
    for batch,(x, y) in enumerate(X_train_loader):
        # Compute prediction and loss
        pred = model(x)
        loss = loss_fn(pred, y)
 
        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        train_loss.append(loss.item())
        
        train_accuracy.append((pred.argmax(1) == y).sum().item() / len(pred))        
        
    test_loss = []
    test_accuracy = []
    model.eval() #turns off dropout
    for i, (x, y) in enumerate(X_test_loader):
        # pass data through network
        pred = model(x)
        loss = loss_fn(pred, y)
        test_loss.append(loss.item())
        test_accuracy.append((pred.argmax(1) == y).sum().item() / len(pred))
 
    print(f'epoch: {t}, train loss: {np.mean(train_loss):.5f}, test loss: {np.mean(test_loss):.5f}, ' +\
          f'train_score: {np.mean(train_accuracy):.1%}, test_score: {np.mean(test_accuracy):.1%}')
    scores.append((np.mean(train_loss),np.mean(test_loss),np.mean(train_accuracy),np.mean(test_accuracy)))
```

```python
#utility plotting function

def show_scores(scores):
    scores = np.array(scores)
    plt.figure(figsize=(12,4))
    plt.subplot(121)
    plt.plot(scores.T[0], '.-', label='train loss')
    plt.plot(scores.T[1], '.-', label='test loss')
    plt.xticks(range(len(scores)), np.arange(len(scores)) + 1)
    plt.legend()
    plt.grid(ls=':');

    plt.subplot(122)
    plt.plot(scores.T[2], '.-', label='train score')
    plt.plot(scores.T[3], '.-', label='test score')
    plt.xticks(range(len(scores)), np.arange(len(scores)) + 1)
    plt.legend()
    plt.grid(ls=':');
```

```python
dropouts = (0,.2,.3,.4,.5,.75,.99) #try all of these
scores2b = defaultdict(list)

model2b = {p:
    nn.Sequential(
        nn.Conv2d(1, 8, kernel_size=(4,4)),
        nn.ReLU(),
    
        nn.Flatten(),
        nn.Dropout(p),
    
        nn.Linear(5000,10),   # connects to 10 node output layer
    )
    for p in dropouts}
           
learning_rate = 1e-3;
loss_fn = nn.CrossEntropyLoss()
```

```python
def doplot():
    for p in dropouts:
       scores = np.array(scores2b[p])
       plt.plot(scores.T[3], '.-', label=p)
    plt.xlabel('epoch')
    plt.ylabel('test score')
    plt.grid('on', ls=':')
    plt.xticks(range(0,20))
    plt.legend();

plt.figure(figsize=(12,4))
plt.subplot(121)
doplot()
plt.subplot(122)
plt.title('(zoomed in)')
doplot()
plt.ylim(.976,.986);
```
