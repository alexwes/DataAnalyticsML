```python
import nltk   #might need `conda install nltk`
# see note above
from nltk.corpus import gutenberg
```

```python
for fileid in gutenberg.fileids():
    print (fileid, len(gutenberg.words(fileid)))
```

```python
#Note: contains punctuation
print (gutenberg.words('milton-paradise.txt')[:20])
```

```python
shf = [fileid for fileid in gutenberg.fileids() if fileid.startswith('shakespeare')]
shf  #shakespeare files
```

```python
shw = [w.lower() for fileid in shf for w in gutenberg.words(fileid) 
       if w[0].isalpha()]
# shakespeare words, punctuation removed;
mpw = [w.lower() for w in gutenberg.words('milton-paradise.txt') if w[0].isalpha()]
len(shw),len(mpw) #numbers of words
```

```python
print ('shakespeare:', Counter(shw).most_common(10))
print ('milton:', Counter(mpw).most_common(10))
```

```python
fdist = Counter(shw + mpw) #for both shakespeare and milton
top50_2,_ = zip(*fdist.most_common(50)) #just use top 50 from combined both
print (top50_2)
```

Now break up the first 65000 words in Shakespeare into 13 blocks of 5000 words apiece, and the first 80000 words of Milton into 16 blocks of 5000 words apiece, and in each block count the number of stopwords:

```python
M2=[]
for corp in [shw, mpw]:
    for i in range(0,len(corp)-4999,5000):  # 13 blocks of shakespeare, 16 of milton
        fdist = Counter(corp[i:i+5000])
        M2.append([fdist[w] for w in top50_2])
M2 = np.array(M2).astype(np.float64) #convert integers to floating point array
```

```python
M2[0] #first 5000 word block of shakespeare
```

Here's what they look like, with each of the 13 blue lines representing one of the 13 blocks of Shakespeare text, and each of the 16 orange lines a block from Milton:

```python
def plot_format(ylab=''):
    plt.xticks(range(50), top50_2, rotation=270)
    plt.xlim(-1,50)
    plt.grid(axis='x', linestyle='dotted')
    plt.ylabel(ylab)
    plt.legend()

plt.figure(figsize=(10,8))

plt.subplot(211)
plt.plot(M2[:13].T, 'C0') #shakespeare blocks in blue
plt.plot(M2[13:].T, 'C1', alpha=.8) #milton blocks in orange
plt.plot([-1],[100],label='Shakespeare (13)') # faux point for legend
plt.plot([-1],[100],label='Milton (16)')      # faux point for legend
plot_format('counts per 5000 word block')

plt.subplot(212)
plt.plot(M2[:13].mean(0), '.-', label='Shakespeare') #shakespeare blocks in blue
plt.plot(M2[13:].mean(0), '.-', label='Milton') #milton blocks in orange
plot_format('means')
```

We're only interested in the deviations from the means, so we can subtract off the mean for each word from its column:

    M2 = M2 - M2.mean(0)[None,:]  # substract off means from columns

and to make all 50 words equally important, we'll divide each column by its standard deviation:

    M2 = M2/M2.std(0) #divide columns by standard deviation
    
Or we can use sklearn.processing.[StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html)() to do the equivalent for us:

```python
scaler2 = preprocessing.StandardScaler()
M2_scaled = scaler2.fit_transform(M2)
```

Now the values for all words have zero mean and similar sized fluctuations about the mean:

```python
plt.figure(figsize=(10,3))
plt.plot(M2_scaled[:13].T, 'C0', alpha=.5)
plt.plot(M2_scaled[13:].T, 'C1', alpha=.5)
plt.plot(M2_scaled[:13].mean(0),'o-', lw=5, label='Sh\'peare mean')
plt.plot(M2_scaled[13:].mean(0),'o-', lw=5, color='#ffdf6e', label='Milton mean')
plt.grid(axis='x')
plt.legend(loc='upper right', fontsize=9)
plt.xticks(range(50), top50_2, rotation=270);
```

Now dimensionally reduce from 50 features to just 2, using [sklearn.decomposition.PCA()](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html):

```python
pca2 = PCA(n_components=2)  #find two
M2_new = pca2.fit_transform(M2_scaled)
evr = pca2.explained_variance_ratio_
print ('explained variance ratio:', evr, sum(evr))  #eigenvalues, scaled to sum to 1
```

Thus the first two principal components account for 53% of the variation of the original 50-dimensional data.<br>
(Note that `PCA(n_components=2)` calculates all 50 diagonal variances of the original 50 dimensional data, even though it only returns the top ones specified by `n_components`. So the "explained variance" ratios above are with respect to the original data, with the sum of all 50 in the denominator.)

And the first component (x-axis) gives a clear separation between the two authors:

```python
plt.scatter(*M2_new.T, s=25, c=['C0']*13 + ['C1']*16)
plt.ylim(-5,5)
plt.title('     Milton                                     Shakespeare');
```

```python
pca2.components_[0]
```

Order these weights from most negative to most positive, and see the corresponding words, with the first few indicative of Shakespeare, and the last few Milton.

```python
print(np.array(top50_2)[pca2.components_[0].argsort()])
```

The separation is good, so try training a k-means classifier using from `sklearn.cluster.KMeans()`:

```python
km2=KMeans(2, random_state=4).fit(M2_new)
means2=km2.cluster_centers_
```

```python
# means2[0] is clearly Milton, means2[1] is Shakespeare:
means2
```

For fun, try classifying the last 5000 word blocks from the two (haven't already seen all of those):

```python
test=[]

tdist = Counter(shw[-5000:])  #last 5000 words of Shakespeare
test.append([tdist[w] for w in top50_2])

tdist = Counter(mpw[-5000:])  #last 5000 words of Milton
test.append([tdist[w] for w in top50_2])

#Do same transform as for training data
test = pca2.transform(scaler2.transform(test))

print(km2.predict(test))  #expect 1 for Shakespeare, then 0 for the Milton block
```

```python
plt.scatter(*M2_new.T, s=25, c=['C0']*13 + ['C1']*16)
plt.ylim(-5,5)
#first principal component gives good separation
plt.plot(*means2.T,'*g',markersize=15, alpha=.8, label='means') #means
plt.plot(*test.T,'o',color='cyan', label='test')               #test
plt.legend(loc='lower center')
plt.title('     Milton                                     Shakespeare');
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#Here's a sample 3d plot
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure(figsize=(5,5))
ax = fig.add_subplot(111, projection='3d', elev=30, azim=-60)

h = np.sqrt(3)/2  #height of equilateral triangle
h3 = np.sqrt(2/3) #height of tetrahedron
v = [(1/2,0,0), (-1/2,0,0), (0,h,0), (0,h/3,h3)]
v -= np.array((0,h/3,h3/4))  #put centroid at origin

plt.plot(*zip(*v),'o-')             #blue
plt.plot(*zip(v[1],v[3],v[0],v[2])) #orange

ax.set(xlim = (-.5, .5), ylim = (h/6-.5, h/6+.5), zlim = (h3/4-.7, h3/4+.3),
       box_aspect = (1,1,1));
```

```python
#check that all six edges are unit length
for i in range(4):
    for j in range(i+1,4):
       print ('|v[{}] - v[{}]| ='.format(i,j), np.linalg.norm(v[i] - v[j]))
```

```python
#now a sample animation (optional)
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
fig=plt.figure(figsize=(5,5))
ax = fig.add_subplot(111, projection='3d')
ax.set_box_aspect((1,1,1))

def animate(f):  #takes frame num as arg
    if f>0: ax.clear()
    ax.plot(*zip(*v),'o-', c='C0') #blue
    ax.plot(*zip(v[1],v[3],v[0],v[2]), c='C1') #orange
    ax.azim=-69+3*f
    
    ax.set(xlim = (-.5, .5), ylim = (h/6-.5, h/6+.5), zlim = (h3/4-.7, h3/4+.3),
             box_aspect = (1,1,1))
    
# call the animator.  blit=True means only re-draw the parts that have changed.
ani = FuncAnimation(fig, animate, #init_func=init,
                               frames=9, interval=200, blit=False)

plt.close(fig)
HTML(ani.to_jshtml(default_mode='reflect'))
```
