```python
nvocab = Counter()
for txt in negtexts:
    for w in set(words(txt)): nvocab[w] += 1
        
pvocab = Counter()
for txt in postexts:
    for w in set(words(txt)): pvocab[w] += 1

allwords = pvocab + nvocab
allwords.most_common(10)
```

```python
Nf=1000
vocab = sorted (allwords, key=allwords.get, reverse=True)[:Nf]  #Nf most common
word_index = {w:k for k,w in enumerate(vocab)} #assigns word each an index by rank
set_vocab = set(vocab)

def features(txt,Nf):
    f = np.zeros(Nf)
    for w in set(words(txt)) & set_vocab: f[word_index[w]] = 1
    return f  #array of zeros and ones for whether or not feature words occur
```

```python
# vocab uses words contained in at least 116 docs
for w in vocab[-10:]: print (w,allwords[w])
```

```python
#http://scikit-learn.org/stable/tutorial/statistical_inference/model_selection.html
#http://scikit-learn.org/stable/modules/cross_validation.html   
#10 fold cross validation
X_data = [features(txt,Nf) for txt in negtexts+postexts]
y_data = [0]*1000 + [1]*1000
```

```python
# i) "lucky" split scores .835
X_train = X_data[:600] + X_data[700:1600] + X_data[1700:]
X_test = X_data[600:700] + X_data[1600:1700]
y_train = [0]*900 + [1]*900
y_test = [0]*100 + [1]*100

clf = BernoulliNB(alpha=.5)
clf.fit(X_train, y_train) #fit the training data
clf.score(X_test, y_test)
```

```python
# ii) "unlucky" split scores .75
X_train = X_data[:300] + X_data[400:1300] + X_data[1400:]
X_test = X_data[300:400] + X_data[1300:1400]

clf = BernoulliNB(alpha=.5)
clf.fit(X_train, y_train) #fit the training data
clf.score(X_test, y_test)
```

```python
# iii) vocab of top 1000 words gives average of 79.35%
scores = cross_val_score(clf, X_data, y_data, cv=10) # or try cv = 20 for 95%/5% splits
scores, scores.mean()
```

```python
#get training scores as well
output = cross_validate(clf, X_data, y_data, return_train_score=True, cv=10)
```

```python
plt.plot(range(10),output['train_score'], 'o-', label='train scores')
plt.plot(range(10),output['test_score'], 'o-', label='test scores')
plt.hlines([output['test_score'].mean()],0,9, linestyle='dotted', color='C1')
plt.title('10-fold validation, BernoulliNB')
plt.xticks(range(10))
plt.grid(axis='y', linestyle='dotted')
plt.ylim(.7,1)
plt.legend();
```

```python
import re, matplotlib.pyplot as plt, numpy as np
%matplotlib inline
from collections import Counter, defaultdict
from sklearn.naive_bayes import BernoulliNB, MultinomialNB
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, plot_confusion_matrix, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import RepeatedStratifiedKFold
```

```python
from ps1data import absdata
classnames = sorted(absdata)
print (classnames)
```

```python
#def words(txt): return txt.lower().split()
#regular expression works better to tokenize and strip punctuation,
# see https://docs.python.org/3/library/re.html
def words(txt): return re.findall(r"['\w]+", txt.lower())
```

```python
#have to fix this to use all unless needed for something else?!?

vocab = {subj: Counter() for subj in classnames}
for subj in classnames:
  for txt in absdata[subj]:
     for w in set(words(txt)): vocab[subj][w] += 1
```

```python
#make a combined total_vocab
total_vocab = Counter()
for subj in classnames: total_vocab += vocab[subj]
len(total_vocab)
```

**A.** Again using `BernoulliNB(alpha=.5)` (though as you can verify, it turns out alpha has little effect on this or the previous problem), implement a twelve way classifier for all twelve classes in problem 2 of ps1, 10-fold cross-validated as above, here on ten splits of 900 abstracts in each of the categories (a total of 10800 documents for each split), and tested on 100 from each of those twelve categories (a total of 1200 documents for each split). 

i) As feature set, use the `Nf = 2500` words contained in the highest percentage of the 12000 documents, excluding the ones that occur in more than 2700 of them (22.5%). Print the list of most frequent words that you've excluded (I found there were 36 of them, so used the next 2500, up to the 2536th, as the 2500 features)

ii) What is the mean score over the 10-fold cross-validation? (You should find somewhere in the mid 80% range.)
Plot the test and training scores as a function of the split, as in 1.A.iii).

```python
# i)
total_vocab.most_common(36)
```

```python
# take Nf most common terms as features, eliminate first 36
Nf = 2500
Nfvocab, Nfn = zip(*total_vocab.most_common(Nf+36)[36:])
word_index = {w:k for k,w in enumerate(Nfvocab)} #assigns each word an index by rank
set_vocab = set(Nfvocab)
```

```python
def features(txt, Nf):
    f = np.zeros(Nf)
    for w in set(words(txt)) & set_vocab: f[word_index[w]] = 1
    return f  #array of zeros and ones for whether or not feature words occur
```

```python
X_data = np.array([features(txt, Nf) for subj in classnames for txt in absdata[subj]]) #all data
y_data = np.array([[i]*1000 for i in range(12)]).flatten() #all data
```

```python
clf = BernoulliNB(alpha=.5)
scores = cross_val_score(clf, X_data, y_data, cv=10) # or try cv = 20 for 95%/5% splits
print ('mean = {:.1%}'.format(round(scores.mean(),3)))
scores
```

**B.** i) As in [lec7_confusion.ipynb](https://nbviewer.jupyter.org/url/courses.cit.cornell.edu/info3950_2021sp/lec7_confusion.ipynb), print the `classification_report()` and use [plot_confusion_matrix](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.plot_confusion_matrix.html) to plot the (12x12) confusion matrix for the first split of the data (training set comprised of first 900 from all of the twelve classes, test set the latter 100 from all of them).

ii) Plot the confusion matrix averaged over the ten cross-validation splits (again as in [lec7_confusion.ipynb](https://nbviewer.jupyter.org/url/courses.cit.cornell.edu/info3950_2021sp/lec7_confusion.ipynb)).

```python
X_test = []
X_train = []
for subj in classnames:
    X_train += [features(txt,Nf) for txt in absdata[subj][:900]]
    X_test += [features(txt,Nf) for txt in absdata[subj][900:]]
    
y_train = np.array([[i]*900 for i in range(12)]).flatten()
y_test = np.array([[i]*100 for i in range(12)]).flatten()
```

```python
clf = BernoulliNB(alpha=.5)
clf.fit(X_train, y_train) #fit the training data
clf.score(X_test, y_test)
```

```python
# i)
print (classification_report(y_test, clf.predict(X_test)))
```

```python
ConfusionMatrixDisplay.from_estimator(clf, X_test, y_test, #ax= so that it uses the above figsize
    display_labels=classnames, xticks_rotation='vertical');
```

```python
clf = BernoulliNB(alpha=.5)
rskf = RepeatedStratifiedKFold(n_splits=10, n_repeats=1) #also ran n_repeats=20 in class
cmat=np.zeros([12,12])

for train_index, test_index in rskf.split(X_data, y_data):
#    print("TRAIN:", train_index, "TEST:", test_index) #indices of the split
    X_train, X_test = X_data[train_index], X_data[test_index]
    y_train, y_test = y_data[train_index], y_data[test_index]
    clf.fit(X_train, y_train) #note that .fit() always reinitializes starts from scratch
    print(clf.score(X_test, y_test)) #just to see the overall scores
    cmat += confusion_matrix(y_test, clf.predict(X_test))
    
cmat /= 10 # mean of the ten splits
```

```python
plt.figure(figsize=(6,6))
ConfusionMatrixDisplay(cmat, display_labels=classnames).plot(ax=plt.gca(),
                                xticks_rotation='vertical');
```

**C.** Use the same 2500 word vocabulary as in part A (it still makes sense to select words according to the percentage of documents in which they occur), and determine whether or not the mean 10-fold validation test score improves.

```python
def nfeatures(txt,Nf):
    f = np.zeros(Nf)
    count = Counter(words(txt))
    for w in set(count) & set_vocab: f[word_index[w]] = count[w]
    return f  #array of number of times each feature occurs in txt
```

```python
X_data = np.array([nfeatures(txt,Nf) for subj in classnames for txt in absdata[subj]]) #all data
y_data = np.array([[i]*1000 for i in range(12)]).flatten() #all data
```

```python
clf = MultinomialNB()  #now defaults to alpha = 1
scores = cross_val_score(clf, X_data, y_data, cv=10) # or try cv = 20 for 95%/5% splits
print ('mean = {:.1%}'.format(round(scores.mean(),3)))
scores
```
