```python
import numpy as np, matplotlib.pyplot as plt
%matplotlib inline
cmap=plt.get_cmap('coolwarm')
arropts = {'width':.001, 'head_width':0.14, 'head_length':0.10, 'length_includes_head':True}
c = cmap([.2,.8])
```

```python
w = np.array([-.5,1])
```

```python
xr = np.array([-2,2])
def plot(b=0):
    plt.xlim(-2,2)
    plt.ylim(-2,2)
    yr = -(w[0]*xr +b)/w[1]
    v = -b*w/w.dot(w)
    plt.fill_between(xr,yr,2,color=c[1],label='bio')
    plt.fill_between(xr,yr,-2,color=c[0],label='phys')
    plt.arrow(v[0], v[1], w[0], w[1], color='#990000', **arropts)
    plt.text(*(v+w),s=r'$\vec w$', ha='center', va='bottom', fontsize=15, color='#990000')
    plt.plot(0,0,'ok',mfc='none')
    plt.plot([0,1,1],[1,0,1],'ok')
    plt.text(1,.1,'gravity',ha='center')
    plt.text(0,1.1,'cell',ha='center')
    plt.text(1,1.1,'gravity,cell',ha='center')
    plt.xticks(range(-2,3))
    plt.yticks(range(-2,3))
    plt.title('$w = (-.5, 1)$, $b = {}$'.format(b))
    plt.xlabel('$x_0$')
    plt.ylabel('$x_1$')
    plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(17,5))
plt.subplot(1,3,1)
plot(0)
plt.subplot(1,3,2)
plot(-.75)
plt.subplot(1,3,3)
plot(.75)
```

The figure above shows a simplified version of a Naive Bayes classifier with a two word vocabulary: 'gravity' and 'cell'. $f(x)=\vec w\cdot \vec x +b >0$ is assumed to indicate biology, and $<0$ physics.
The weights are $\vec w = (-.5, 1)$, meaning that 'gravity' signifies the physics (negative) direction, and 'cell' is twice as important in the biology (positive) direction.

The decision boundary for $\vec w\cdot \vec x +b$ is shown for three values of the bias $b$. The three 'documents' plotted consist of the word ('gravity'), ('cell'), and ('gravity', 'cell').

In the plot at left, with decision boundary given by $-.5x_0 +x_1 = 0$, the upper two documents are classified as biology (red background), and the lower as physics (blue background).

In the middle plot, with decision boundary given by $-.5x_0 +x_1 -.75 = 0$, the decision boundary moves upward (negative bias makes it harder for $f(x)>0$), and only the document ('cell') is classified as biology.

In the plot at right, with decision boundary given by $-.5x_0 +x_1 + .75 = 0$, the decision boundary moves downward (positive bias makes it easier for $f(x)>0$), and all three documents are classified as biology.
