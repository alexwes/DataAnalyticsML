```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
failure=[70, 57, 63, 70, 53, 75, 58]
success=[66, 69, 68, 67, 72, 73, 70, 78, 67, 67, 75, 70, 81, 76, 79, 76]
xdata,ydata=zip(*[[t,0] for t in failure] + [[t,1] for t in success])

X = np.array(xdata)[:, None]
y = np.array(ydata)
```

```python
from sklearn import linear_model
clf = linear_model.LogisticRegression(C=1e6, tol=1e-6).fit(X, y) #turn off regulation
```

```python
a = clf.coef_[0][0]
t0 = -clf.intercept_[0]/a
print ('a={:.3f}, t0={:.2f}'.format(a,t0))
```

```python
def p(t,a=1.,t0=0.): return 1./(1.+np.exp(-a*(t-t0)))
#calculates log(prob), using 1-p for points with success
def logprob(a,t0):
   data_probs = np.array([p(t,a,t0) if result else 1-p(t,a,t0) for t,result in zip(xdata,ydata)])
   return np.sum(np.log(data_probs))
```

```python
#these values give a slightly larger probability, better fit to data
logprob(.232, 64.79), logprob(.24, 65)
```

```python
#B.
t0r = np.linspace(t0-2.5,t0+2.5,100)
ar = np.linspace(a-.025, a+.025, 100)
logprobs = np.array([[logprob(a,t0) for t0 in t0r] for a in ar])
pprobs = np.exp(logprobs) / np.exp(logprobs).sum()  #now consider normalized posterior probs
```

```python
#look directly at the overall probability of the data for all values of t0, a
from matplotlib.lines import Line2D #to make custom legend handles
plt.figure(figsize=(7, 6))

plt.contourf(t0r, ar, pprobs, cmap='jet')
plt.grid('on', color='w', linestyle='dotted')
plt.title('zoomed near new max (a={:.3f}, t0={:.2f})'.format(a, t0))
plt.xlabel('t0'), plt.ylabel('a')
plt.colorbar()
plt.scatter([65, t0], [.24, a], color=['g','k'])
plt.legend([Line2D([0], [0], marker='o', c='none', mfc=c, mec='none', ms=7) for c in 'gk'],
           ['earlier grid max', 'more accurate max']);
```

```python
#C.
clf.score(X,y) # so 20/23 correct
```

```python
# it's the 3 red Failure points above 65
X[clf.predict(X) != y].flatten(), y[clf.predict(X) != y]
```

```python
import numpy as np, matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets, linear_model
digits = datasets.load_digits()
X_digits = digits.data
y_digits = digits.target

def unrav(k): return (k%8-.5, k//8-.5) #or use np.unravel_index() to translate 
```

### 2.A

```python
n_samples = len(X_digits)
n_training=int(.9*n_samples)
print ('n_training={}, n_test={}'.format(n_training, n_samples-n_training))
X_train = X_digits[:n_training] #first 90%
y_train = y_digits[:n_training]
X_test = X_digits[n_training:]  #last 10%
y_test = y_digits[n_training:]

logreg_ovr = linear_model.LogisticRegression(multi_class='ovr', max_iter=1000).fit(X_train, y_train)
```

```python
print('LogisticRegression score (ovr): {:.6f}'.format(logreg_ovr.score(X_test, y_test)))
print ('169/180 = {:.6f}, so 11 wrong'.format(169/180))
```

```python
n_training=int(.9*n_samples)
n_samples-n_training
```

```python
y_pred = logreg_ovr.predict(X_test)
diffs = y_pred != y_test

print ('    indices:', np.where(diffs)[0])
print ('predictions:', y_pred[diffs])
print ('    actuals:', y_test[diffs])
```

```python
fig,axi = plt.subplots(2,6,figsize=(12, 4))
for i,ax in zip(np.where(y_pred != y_test)[0],axi.flatten()):
    ax.imshow(X_test[i].reshape(8,8),cmap='Greys', aspect=1, interpolation='none')
    ax.set_xticks([]), ax.set_yticks([])
    ax.set_title(f'{i}: pred={y_pred[i]}, true={y_test[i]}',fontsize=10,color='b')
axi[-1,-1].axis(False);
```
