```python
# see s6_supp.ipynb for code to save the data from ps5
M5_pca, y5_train, T5_pca, y5_test, scaler5, pca5 = np.load('5auths.npy', allow_pickle=True)
```

```python
from matplotlib.colors import LinearSegmentedColormap
C02386 = LinearSegmentedColormap.from_list('C02386', 'C0 C2 C3 C8 C6'.split(), 5)
clrs5 = np.array(['b', 'g', 'r', 'y', 'C6'])
```

```python
# utility plot
xx, yy = np.meshgrid(np.linspace(-7.5, 9.5, 200), np.linspace(-8, 9, 200))

def mkfig(clf, ax, show_test=False, show_legend=False):
    A = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    A = A.max(1).reshape(xx.shape)
    B = (A-A.min()) / (A.max()-A.min()) #for the alpha

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.scatter(*M5_pca.T[:2], zorder=2, c=clrs5[y5_train], ec='k')
    tr_score = clf.score(M5_pca[:,:2],y5_train)
    if show_test:
        te_score = clf.score(T5_pca[:,:2],y5_test)
        ax.scatter(*T5_pca.T[:2], zorder=2, c=clrs5[y5_test], ec='k', marker='d')
        ax.text(-7,-9,f'train={tr_score:.1%}, test={te_score:.1%}')
    else:
        ax.text(-3,-9,f'train={tr_score:.1%}') # just train score
    ax.set_aspect(1)
    ax.pcolormesh(xx, yy, Z, cmap=C02386, alpha=B)
    ax.axis('off')
    if show_legend: mylegend(ax);
    
from matplotlib.lines import Line2D #to make custom legend handles
def lmarker(c, m='o', ms=7): return Line2D([0], [0], marker=m, c='none', mfc=c, mec='none', ms=ms)
auths = ['shakespeare', 'milton', 'austen', 'melville', 'christie']

def mylegend(ax, loc='upper right'):
    return ax.legend([lmarker(c) for c in clrs5], auths, loc=loc, fontsize=8)
```

```python
# default SVC(), C=1, gamma='scaled'
clf = SVC(probability=True)
clf.fit(M5_pca[:,:2], y5_train)
print ('default', clf.score(M5_pca[:,:2],y5_train), clf.score(T5_pca[:,:2],y5_test))
```

```python
# have a look
fig, ax = plt.subplots(1,2,figsize=(8,4))
mkfig(clf, ax[0], show_legend=True)
mkfig(clf, ax[1], show_test=True, show_legend=True)
```

Note that this is something of an unprincipled test, since we're projecting down to two dimensions and the third dimension will separate austen and christie

```python
Cg = [(C,gamma) for C in (.1,1,10) for gamma in (.015, .1, 1)]

fig,ax = plt.subplots(3,3, figsize=(8, 8))
for axi,(C,gamma) in zip(ax.flatten(),Cg):
    clf = SVC(probability=True, C=C, gamma=gamma)
    clf.fit(M5_pca[:,:2], y5_train)
    mkfig(clf, axi, show_test=True)
    if axi.get_subplotspec().is_first_row(): axi.set_title(f'gamma={gamma}')
    if axi.get_subplotspec().is_first_col(): axi.text(-12.5,0, f'C={C}', fontsize=12)
```

```python
#here's another test = 91%
clf2 = SVC(C=.85, gamma=.223, probability=True)
clf2.fit(M5_pca[:,:2], y5_train)

fig, ax = plt.subplots(1,2,figsize=(8,4))
mkfig(clf2, ax[0], show_legend=True)
mkfig(clf2, ax[1], show_test=True, show_legend=True)
```

# 2.

```python
allcount = Counter()
for auth in sorted(glob('ps6data/train/*/')):
    for f in sorted(glob(f'{auth}/*')):
        txt = open(f).read()
        allcount += Counter([txt[i:i+3] for i in range(len(txt)-2)])
```

```python
# top 4000 trigrams
top4000,_ = zip(*allcount.most_common(4000)) #we'll only use first 1500 here, rest for part B
```

```python
top4000[:10]
```

```python
# ingest the train set
X_train = []
y_train = []
for auth in sorted(glob('ps6data/train/*')):
    for f in sorted(glob(f'{auth}/*')):
        txt = open(f).read()
        c = Counter([txt[i:i+3] for i in range(len(txt)-2)])
        X_train.append([c[tg]/len(txt) for tg in top4000])
        y_train.append(auth.split('/')[-1])
        
X_train = np.array(X_train)
y_train = np.array(y_train)
X_train.shape, y_train.shape
```

```python
# ingest the test set
X_test = []
y_test = []
for auth in sorted(glob('ps6data/test/*')):
    for f in sorted(glob(f'{auth}/*')):
        txt = open(f).read()
        c = Counter([txt[i:i+3] for i in range(len(txt)-2)])
        X_test.append([c[tg]/len(txt) for tg in top4000])
        y_test.append(auth.split('/')[-1])

X_test = np.array(X_test)
y_test = np.array(y_test)
X_test.shape, y_test.shape
```

```python
scaler = preprocessing.StandardScaler()
nf = 1500 # this reduces number of features for training

X_train_scaled = scaler.fit_transform(X_train[:,:nf])
X_test_scaled = scaler.transform(X_test[:,:nf])
X_train_scaled.shape, X_test_scaled.shape
```

## 2A

```python
# first the default C,gamma
clf = SVC()
cv8 = cross_val_score(clf, X_train_scaled, y_train, cv=8)
print (nf, f'mean={cv8.mean()}')
```

```python
# can do better (would also do better with 2000 features ...)
clf = SVC(C=6, gamma=7e-05)
cv8 = cross_val_score(clf, X_train_scaled, y_train, cv=8)
print (f'mean={cv8.mean()}')
```

## 2A.i

```python
%%time

gamma_range = np.arange(.00001,.0001,.00001)
C_range = np.arange(2,10)
param_grid = {'C': C_range, 'gamma':  gamma_range}

clf = GridSearchCV(SVC(kernel='rbf', class_weight='balanced'),
                   param_grid, cv=8, verbose=True)

clf.fit(X_train_scaled, y_train)
```

```python
# note that this will depend on the random state
clf.best_estimator_, clf.best_params_, clf.best_score_
```

```python
scores = clf.cv_results_['mean_test_score'].reshape(len(C_range),len(gamma_range))
```

```python
plt.imshow(scores, norm=TwoSlopeNorm(vcenter=0.9, vmax=.91)) #note that TwoSlopeNorm fixes earlier bugs
plt.xlabel("gamma")
plt.ylabel("C")
plt.colorbar(ticks = list(np.arange(.82,.911,.02)) + [.905,.91])
plt.xticks(np.arange(len(gamma_range)), [f'{x:g}' for x in gamma_range], rotation=45)
plt.yticks(np.arange(len(C_range)), [f'{x:g}' for x in C_range])
plt.title("Validation accuracy");
```

```python
auths = sorted(set(y_train))
rskf = RepeatedStratifiedKFold(n_splits=8, n_repeats=1)
cmat=np.zeros([20,20])

for train_index, test_index in rskf.split(X_train_scaled, y_train):
    X_tr, X_te = X_train_scaled[train_index], X_train_scaled[test_index]
    y_tr, y_te = y_train[train_index], y_train[test_index]
    clf.best_estimator_.fit(X_tr, y_tr) #always reinitializes
    print(clf.score(X_te, y_te), end=' ') #just to see the overall scores
    cmat += confusion_matrix(y_te, clf.predict(X_te))
cmat /= 8
```

```python
plt.figure(figsize=(9,9))
ConfusionMatrixDisplay(cmat, display_labels=auths).plot(ax=plt.gca(), cmap='Reds',
                                 values_format='.1f');
```

fb predicted to be cb, and bs predicted to be mg. both have value .375, so wrong in that way 3 times in the 8 cross validations

(I may have mislead some into plotting the overall train and test for this -- since this would be used to decide the number of features it would have to be done only on the train/validation set. But below is an equivalent plot for training/test features.)

```python
nfs = [1,10,25,50,75,100,250,500,750,1000,1500,2000,3000,4000]
scores= []
for nf in nfs:
    X_train_scaled = scaler.fit_transform(X_train[:,:nf])
    X_test_scaled = scaler.transform(X_test[:,:nf])
        
    clfb.fit(X_train_scaled,y_train)
    scores.append((clfb.score(X_train_scaled,y_train), clfb.score(X_test_scaled,y_test)))
    print (nf, scores[-1])
    
scores = np.array(scores)
```

```python
plt.plot(nfs, scores.T[0], '.-', label='train score')
plt.plot(nfs, scores.T[1], '.-', label='test score')
plt.xlabel('# features')
plt.yticks(np.arange(0,1.1,.1))
plt.grid(ls=':')
plt.legend(loc='center right');
```
